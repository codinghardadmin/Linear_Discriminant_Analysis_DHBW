
# Comparison of LDA and PCA 2D projection of Iris dataset


The Iris dataset represents 3 kind of Iris flowers (Setosa, Versicolour
and Virginica) with 4 attributes: sepal length, sepal width, petal length
and petal width.

Principal Component Analysis (PCA) applied to this data identifies the
combination of attributes (principal components, or directions in the
feature space) that account for the most variance in the data. Here we
plot the different samples on the 2 first principal components.

Linear Discriminant Analysis (LDA) tries to identify attributes that
account for the most variance *between classes*. In particular,
LDA, in contrast to PCA, is a supervised method, using known class labels.


```python
print(__doc__)

import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

print("###################")
print(target_names)
print(y)

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')

# plt.show()

print("#####################")
print("#####################")

# _lda

# print(lda.predict([0.5, 0.5, 0.5]))

print(X_r2)
```

# Eigener Test

Eigener Test für LDA

```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt

# Daten erstellen
X = np.array([[-10, -1, 2], [-20, -1, 2], [-15, -2, 2], [10, 1, 2], [20, 1, 2], [15, 2, 2], [15, 2, -10]])

# Zuordnung (überwachtes Lernen)
y = np.array([1, 1, 1, 2, 2, 2, 3])

# Linear Discriminant Analysis
clf = LinearDiscriminantAnalysis()

# Daten eingeben
clf.fit(X, y)
clf_2 = clf.transform(X)

print(clf.get_params())
print(clf_2)


#LinearDiscriminantAnalysis()
print("Predict:")
print(clf.predict([[30.8, -1, 2]]))
print("Predict_Proba:")
print(clf.predict_proba([[30.8, -10, 2]]))
print("Predict_Log_Proba:")
print(clf.predict_log_proba([[30.8, -10, 2]]))

print("Predict_Log_Proba:")
print(clf.decision_function(X))

# Draw
colors = ['navy', 'turquoise']
lw = 2

plt.figure()
for color, i, target_name in zip(colors, [0, 1], ["k1", "k2"]):
    plt.scatter(clf_2[y == i, 0], clf_2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')

plt.show()
```
